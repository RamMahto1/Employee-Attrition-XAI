```python
## split the data into independent and dependent variable
```

```python
X = df.drop("Attrition",axis=1)
y = df['Attrition']
X.shape, y.shape
```

```python
## split the into train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, random_state=42)
X_train.shape, X_test.shape
```

```python
## feature engineering apply tranformation with is 3 types
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer


numerical_feature = X.select_dtypes(include=['int64']).columns.tolist()
categorical_feature = X.select_dtypes(include=['object']).columns.tolist()



num_pipeline = Pipeline(
    steps=[
        ("num",SimpleImputer(strategy="median")),
        ("scaler",StandardScaler())
    ]
)

cat_pipeline = Pipeline(
    steps=[
        ("cat",SimpleImputer(strategy="most_frequent")),
        ("one_hotencorder",OneHotEncoder(handle_unknown="ignore"))
       
    ]
)


preprocessor = ColumnTransformer(
    [
        ("num_pipe",num_pipeline,numerical_feature),
        ("cat_pipe",cat_pipeline,categorical_feature)
    ]
)
```

```python
X_train_transformer=preprocessor.fit_transform(X_train)
X_test_transformer = preprocessor.transform(X_test)
```

```python
# Check shapes
print("X_train shape:", X_train_transformer.shape)
print("X_test shape:", X_test_transformer.shape)
```

```python
## Apply SMOTE on training data
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
from collections import Counter
```

```python


# Encode target
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)  # 'No'->0, 'Yes'->1
y_test_encoded = le.transform(y_test)        # must use transform

# Apply SMOTE only once on encoded y_train
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_transformer, y_train_encoded)

#  Check new class distribution
print("Original y_train distribution:", Counter(y_train_encoded))
print("Resampled y_train distribution:", Counter(y_train_res))
```

```python
## training the model
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import(
RandomForestClassifier,
AdaBoostClassifier,
GradientBoostingClassifier
)
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

```python
models ={
    "DecisionTreeRegressor":DecisionTreeClassifier(),
    "RandomforestRegressor":RandomForestClassifier(),
    "AdaboostingRegressor":AdaBoostClassifier(),
    "GradietDecentRegressor":GradientBoostingClassifier(),
    "supportVectorClassifier":SVC(),
    "LogisticRegressor":LogisticRegression(),
    "XGBoostingClassifier":XGBClassifier(),
    "CatboostingClassifier":CatBoostClassifier()
}

from sklearn.metrics import precision_score,f1_score,recall_score,accuracy_score,classification_report
```

```python
results = []

for model_name, model in models.items():
    # Train on SMOTE-resampled data
    model.fit(X_train_res, y_train_res)

    # Predictions on original training set
    y_pred_train = model.predict(X_train_transformer)
    # Predictions on test set
    y_pred_test = model.predict(X_test_transformer)

    # Evaluate on training set
    precision_train = precision_score(y_train_encoded, y_pred_train)
    f1_train = f1_score(y_train_encoded, y_pred_train)
    recall_train = recall_score(y_train_encoded, y_pred_train)
    accuracy_train = accuracy_score(y_train_encoded, y_pred_train)

    # Evaluate on test set
    precision_test = precision_score(y_test_encoded, y_pred_test)
    f1_test = f1_score(y_test_encoded, y_pred_test)
    recall_test = recall_score(y_test_encoded, y_pred_test)
    accuracy_test = accuracy_score(y_test_encoded, y_pred_test)

    # Store results
    results.append({
        'Model Name': model_name,
        'precision_score_train': precision_train,
        'f1_score_train': f1_train,
        'recall_score_train': recall_train,
        'accuracy_score_train': accuracy_train,
        'precision_score_test': precision_test,
        'f1_score_test': f1_test,
        'recall_score_test': recall_test,
        'accuracy_score_test': accuracy_test
    })
```

```python
final_result=pd.DataFrame(results)
final_result
```
